## Covertype Dataset

```python
covertype = dfs[4].drop(columns="Unnamed: 0").copy()
```

```python
covertype
```

```python
covertype.info()
```

```python
def save_files(filename:str, df:pd.DataFrame, target_col:str):
    print(100*df[target_col].value_counts()/df.shape[0])
    
    df.to_csv(os.path.join(prep_data_folder,"{}.csv".format(filename)), index=False)
    
    df_train, df_test = train_test_split(df, test_size=0.2, random_state=42)
    
    print("Train class dist")
    print(100*df_train[target_col].value_counts()/df_train.shape[0])
    
    print("Test class dist")
    print(100*df_test[target_col].value_counts()/df_test.shape[0])
    
    df_train.to_csv(os.path.join(train_data_folder,"{}_train.csv".format(filename)), index=False)
    df_test.to_csv(os.path.join(test_data_folder,"{}_test.csv".format(filename)), index=False)
    print("Files saved")
    return df_train, df_test
```

```python
tr, tst = save_files("covertype",covertype, "target")
```
