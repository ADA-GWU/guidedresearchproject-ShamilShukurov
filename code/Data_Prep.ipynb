```python
import pandas as pd
import numpy as np
import sweetviz as sv
from scipy import stats
```

```python
white_wine\
=pd.read_csv(
    r"C:\Users\SukurovSB\Desktop\Guided_Research\guidedresearchproject-ShamilShukurov\data\winequality-white.csv", 
    delimiter=";"
)
```

```python
red_wine.quality.value_counts()/red_wine.quality.value_counts().sum()
```

```python
white_wine.quality.value_counts()/white_wine.quality.value_counts().sum()
```

```python
wine_df = pd.concat([red_wine, white_wine])
```

```python
wine_df["target"] = np.where(wine_df.quality<=4,0,1)
```

```python
wine_df.target.value_counts()/wine_df.target.value_counts().sum() 
```

```python
if "quality" in wine_df.columns:
    wine_df.drop(columns="quality",inplace=True)
```

```python
for col in wine_df.columns[:-1]:
    t = stats.ttest_ind(wine_df[wine_df.target==1][col],wine_df[wine_df.target==0][col])
    print("Column:{}\nt:{}\np_value:{}".format(col,t.statistic.round(4),t.pvalue.round(4)))
    print("===========================")
```
