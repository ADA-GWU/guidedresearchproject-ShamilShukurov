# Letter-A

```python
letter_a_orig = pd.read_csv(join(data_folder,"letter-recognition.data"), header=None, 
                            names= ['lettr','x-box','y-box','width','high ','onpix','x-bar','y-bar','x2bar','y2bar',
                                    'xybar','x2ybr','xy2br','x-ege','xegvy','y-ege','yegvx'])
```

```python
letter_a_orig
```

```python
letter_a_orig["letter_a"]=0
letter_a_orig.loc[letter_a_orig.lettr=='A',"letter_a"]=1
```

```python
#class distribution
100*letter_a_orig.letter_a.value_counts()/letter_a_orig.shape[0]
```

```python
letter_a = letter_a_orig.drop(columns="lettr")
```

```python
letter_a
```

```python
letter_a.to_csv(join(prep_data_folder,"letter_a.csv"))
```

# Letter Vowel

```python
letter_a_orig.lettr.unique()
```

```python
letter_a_orig["is_vowel"]=0
letter_a_orig.loc[letter_a_orig.lettr.isin(['A','E','I','O','U']),"is_vowel"]=1
```

```python
#class distribution
100*letter_a_orig.is_vowel.value_counts()/letter_a_orig.shape[0]
```

```python
letter_vowel = letter_a_orig.drop(columns=["lettr","letter_a"])
```

```python
letter_vowel
```

```python
letter_vowel.to_csv(join(prep_data_folder,"letter_vowel.csv"))
```

# Pendigits

```python
pendigits_orig_test = pd.read_csv(join(data_folder,"pendigits.tes"),header=None,
                            names=['C1','C2','C3','C4','C5','C6','C7','C8','C9','C10','C11','C12','C13','C14','C15','target'])
pendigits_orig_train = pd.read_csv(join(data_folder,"pendigits.tra"),header=None,
                            names=['C1','C2','C3','C4','C5','C6','C7','C8','C9','C10','C11','C12','C13','C14','C15','target'])

pendigits_orig = pd.concat([pendigits_orig_train,pendigits_orig_test])
```

```python
100*pendigits_orig.target.value_counts()/pendigits_orig.shape[0]
```

```python
pendigits_orig["is_9"]=0
pendigits_orig.loc[pendigits_orig.target==9,"is_9"]=1
```

```python
pendigits_orig
```

```python
#class distribution
100*pendigits_orig.is_9.value_counts()/pendigits_orig.shape[0]
```

```python
pendigits = pendigits_orig.drop(columns="target")
```

```python
pendigits
```

```python
pendigits.to_csv(join(prep_data_folder,"pendigits.csv"))
```
