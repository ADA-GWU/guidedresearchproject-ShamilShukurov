# Sick-euthyroid

```python
sick_euthyroid = pd.read_csv(join(data_folder,"sick-euthyroid.data"), header=None,
                            names=['target','age','sex','on_thyroxine','query_on_thyroxine','on_antithyroid_medication',
                                   'thyroid_surgery','query_hypothyroid','query_hyperthyroid','pregnant','sick',
                                   'tumor','lithium','goitre','TSH_measured','TSH','T3_measured','T3','TT4_measured',
                                   'TT4','T4U_measured','T4U','FTI_measured','FTI','TBG_measured','TBG'])
```

```python
sick_euthyroid.target.value_counts()/sick_euthyroid.shape[0]
```

```python
sick_euthyroid.replace('?', np.NaN, inplace=True)
sick_euthyroid.replace('f', 0, inplace=True)
sick_euthyroid.replace('t', 1, inplace=True)
sick_euthyroid.replace('y', 1, inplace=True)
sick_euthyroid.replace('n', 0, inplace=True)
sick_euthyroid.replace('F', 1, inplace=True)
sick_euthyroid.replace('M', 0, inplace=True)
sick_euthyroid.replace('negative', 0, inplace=True)
sick_euthyroid.replace('sick-euthyroid', 1, inplace=True)
```

```python
sick_euthyroid['age'] = pd.to_numeric(sick_euthyroid['age'])
sick_euthyroid.age.fillna(sick_euthyroid.age.median(),inplace=True)
sick_euthyroid.sex.fillna(1,inplace=True)
```

```python
#Missing values
100*sick_euthyroid.isna().sum()/sick_euthyroid.shape[0]
```

```python
sick_euthyroid['TSH'] = pd.to_numeric(sick_euthyroid['TSH'])
sick_euthyroid['TSH'] = pd.to_numeric(sick_euthyroid['TSH'])
sick_euthyroid['TT4'] = pd.to_numeric(sick_euthyroid['TT4'])
sick_euthyroid['T4U'] = pd.to_numeric(sick_euthyroid['T4U'])
sick_euthyroid['FTI'] = pd.to_numeric(sick_euthyroid['FTI'])
sick_euthyroid['TBG'] = pd.to_numeric(sick_euthyroid['TBG'])
```

```python
sick_euthyroid.describe()
```

```python
sick_euthyroid.info()
```

```python
sick_euthyroid.drop(columns=["TSH","T3","TBG"],inplace=True)
```

```python
sick_euthyroid.dropna(inplace=True)
```

```python
#Class distribution
sick_euthyroid.target.value_counts()/sick_euthyroid.shape[0]
```

```python
sick_euthyroid.to_csv(join(prep_data_folder,"sick_euthyroid.csv"))
```
